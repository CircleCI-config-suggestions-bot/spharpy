```python
import numpy as np

%matplotlib notebook
```

```python
import pysh.samplings as samplings
import pysh.plot as shp
from pysh.samplings import sph2cart
```

```python
r, theta, phi, weights = samplings.equiangular(4)

x,y,z = sph2cart(r,theta,phi)
shp.scatter(x,y,z)
```

```python
r, theta, phi, weights = samplings.gaussian(4)

x,y,z = sph2cart(r,theta,phi)
shp.scatter(x,y,z)
```

```python
r, theta, phi = samplings.eigenmike_em32()

x,y,z = sph2cart(r,theta,phi)
shp.scatter(x,y,z)
```

```python
r, theta, phi = samplings.icosahedron()

x,y,z = sph2cart(r,theta,phi)
shp.scatter(x,y,z)
```

```python
n_max = 4
r,theta,phi,w = samplings.hyperinterpolation(n_max)

x,y,z = sph2cart(r,theta,phi)
shp.scatter(x,y,z)
```

```python
n_max = 3

rad, theta, phi = samplings.healpix(n_max)

x,y,z = sph2cart(rad,theta,phi)
shp.scatter(x,y,z)
```
